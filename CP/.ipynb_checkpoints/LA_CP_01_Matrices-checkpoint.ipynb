```python
# these imports may be useful

import pandas as pd
import numpy as np
```

 #### –î–µ—è–∫—ñ —ñ–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∏, —è–∫—ñ, –º–æ–∂–ª–∏–≤–æ, —Å—Ç–∞–Ω—É—Ç—å —É –Ω–∞–≥–æ–¥—ñ.

```python
# these tools may be useful
# give a matrix
a = np.array([[1,2], [2,4], [3,6]]) #3x2 matrix
b = np.array([[1], [2]])            #2x1 matrix = column vector
c = np.array([[1,2], [2,3]])        #2x2 square matrix

#zero matrix
z = np.zeros((2, 3))

#matrix with all 1 elements
ones = np.ones((2,3))

# transpose
at = a.transpose()

# matrix multiplication
prod1 = a @ b
prod2 = np.matmul(a, b)

# dot (inner, scalar) product of vectors
u = np.array([1,2,3])
v = np.array([2,3,4])
uv = np.dot(u, v)

# inverse matrix
c_inv = np.linalg.inv(c)

print(f' zero={z},\n\n all 1-matrix={ones},\n\n a= {a},\n\n b = {b},\n\n,  c= {c}, \n\n a^t = {at},\n\n ab = {prod1}, \n\n ab= {prod2} \n')
print(f'u={u}, v={v}, uv = {uv} \n\n')
print(f'c = {c},\n\n c^(-1) = {c_inv}')
```
